# Data Dictionary

## 0. Load Data

```python
import pandas as pd
import numpy as np
# Load the dataset
df = pd.read_csv('data/raw/cardiovascular_risk_dataset.csv')

# Display the first few rows of the dataset
df.head()
```

## 1. Create Data Dictionary

```python
# Create Data Dictionary Base
data_dictionary = pd.DataFrame({
    'Field Name': ["Patient_ID","age", "bmi", "systolic_bp", "diastolic_bp", "cholesterol_mg_dl",
                   "resting_heart_rate", "smoking_status", "daily_steps", "stress_level",
                   "physical_activity_hours_per_week", "sleep_hours", "family_history_heart_disease",
                   "diet_quality_score", "alcohol_units_per_week", "heart_disease_risk_score", "risk_category"],
})

data_dictionary
```

```python
# Add Data Type
data_dictionary['Data Type'] = data_dictionary['Field Name'].apply(lambda x: str(df[x].dtype))

data_dictionary
```

```python
# Add Description
descriptions = {
    "Patient_ID": "Unique identifier for each patient.",
    "age": "Age of the patient in years.",
    "bmi": "Body Mass Index, indicating body fat based on height and weight.",
    "systolic_bp": "Systolic blood pressure level.",
    "diastolic_bp": "Diastolic blood pressure level.",
    "cholesterol_mg_dl": "Total cholesterol level in mg/dL.",
    "resting_heart_rate": "Resting heart rate in beats per minute.",
    "smoking_status": "Smoking category (Never, Former, Current).",
    "daily_steps": "Average number of steps taken per day.",
    "stress_level": "Stress level on a scale from 1 to 10.",
    "physical_activity_hours_per_week": "Weekly hours of physical exercise.",
    "sleep_hours": "Average hours of sleep per night.",
    "family_history_heart_disease": "Indicates whether the patient has a family history of heart disease.",
    "diet_quality_score": "Score (1–10) representing the quality of diet.",
    "alcohol_units_per_week": "Weekly alcohol consumption.",
    "heart_disease_risk_score": "Calculated cardiovascular risk score (0–100).",
    "risk_category": "Risk classification: Low, Medium, or High."
}

data_dictionary['Description'] = data_dictionary['Field Name'].map(descriptions)

data_dictionary
```

```python
# Add Length
data_dictionary['Length'] = data_dictionary['Field Name'].apply(lambda x: df[x].shape[0])

# Add Unique Values 
data_dictionary['Unique Values'] = data_dictionary['Field Name'].apply(lambda x: df[x].nunique())

#Add Missing Values
data_dictionary['Missing Values'] = data_dictionary['Field Name'].apply(lambda x: df[x].isnull().sum())

data_dictionary
```

```python
# Add Minimum and Maximum Values
def min_max_uniform(col):
    try:
        # Try numeric conversion first
        numeric_col = pd.to_numeric(df[col], errors='coerce')
        if numeric_col.notnull().any():
            return pd.Series({'Min.': numeric_col.min(), 'Max.': numeric_col.max()})
        else:
            # If object, use alphabetical min/max
            obj_col = df[col].dropna().astype(str)
            return pd.Series({'Min.': obj_col.min(), 'Max.': obj_col.max()})
    except:
        return pd.Series({'Min.': np.nan, 'Max.': np.nan})

min_max_df = data_dictionary['Field Name'].apply(min_max_uniform)
data_dictionary = pd.concat([data_dictionary, min_max_df], axis=1)

data_dictionary
```
